## 영화 댓글 크롤링

* 사용방법
> `movie.naver.com` 페이지에 들어가 우측상단에 원하는 영화를 검색한다.
>
> 해당 영화 화면의 URL을 입력!

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 이부분을 바꿔서 사용한다.
영화페이지 = 'https://movie.naver.com/movie/bi/mi/review.nhn?code=150637'
```

```python
tokens = []
# 크롤링할 댓글수 (10 배수)
num = 10 # 수정요망
for i in range(1,num+1):
    url = 영화페이지 + f'&page={i}'
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
        # .class
        # 태그
        # 바로아래 > / 그냥 (띄어쓰기)
    for j in range(10): 
        links =soup.select('.review p > a')[j]['onclick']
        tokens.append(links[-9:-2])
```

```python
# 본 분석 때는 tokens으로 변경
reviews = []
for token in tokens:
    review_url = 'https://movie.naver.com/movie/bi/mi/reviewread.nhn?nid='+token+'&code=150637&order=#tab'
    # print(review_url)
    review_request = requests.get(review_url)
    review_html = review_request.text
    review_soup = BeautifulSoup(review_html, 'html.parser')
    review = review_soup.select('.user_tx_area p')
    
    tmp =''
    for a in range(len(review)):
        tmp += review[a].text

    # 엔터 제거
    tmp = tmp.strip('\r\n')
    tmp = tmp.replace('\xa0',' ')
    reviews.append(tmp)
```

```python
for i in range(len(reviews)):
    print(f'< {i+1}번 댓글 내용 >')
    print('')
    print(reviews[i])
    print('--------------------------------------')    
```
